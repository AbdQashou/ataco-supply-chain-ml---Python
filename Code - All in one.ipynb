# models

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import pandas as pd

# results table
results = pd.DataFrame(columns=['Model','Num_K','Accuracy','Precision','Recall','F1_Score'])

# loop k=3 to 10
for k in range(3, 11):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_cls, y_train_cls)
    y_pred = knn.predict(X_test_cls)

    results.loc[len(results)] = [
        'KNN', k,
        accuracy_score(y_test_cls, y_pred),
        precision_score(y_test_cls, y_pred, average='weighted'),
        recall_score(y_test_cls, y_pred, average='weighted'),
        f1_score(y_test_cls, y_pred, average='weighted')
    ]

print(results)

# best k based onn the accuracy
best_row = results.loc[results['Accuracy'].idxmax()]
best_k = int(best_row['Num_K'])
print(f"Best k = {best_k}, Accuracy = {best_row['Accuracy']:.4f}")
```

```python
# confusion matrix for the best k
best_knn = KNeighborsClassifier(n_neighbors=best_k)
best_knn.fit(X_train_cls, y_train_cls)
y_best_pred = best_knn.predict(X_test_cls)
print("Confusion Matrix:\n", confusion_matrix(y_test_cls, y_best_pred))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(results['Num_K'], results['Accuracy'], marker='o')
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Accuracy')
plt.title('KNN Accuracy vs. Number of Neighbors')
plt.xticks(results['Num_K'])
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix
import numpy as np
```

```python
X = X_cls.copy()
y = y_cls.copy()

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test  = sc.transform(X_test)
```

```python
print("Train shape:", X_train.shape, " Test shape:", X_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(units=32, activation='relu', input_dim=X_train.shape[1]))
model.add(Dense(units=16, activation='relu'))
model.add(Dense(units=1,  activation='sigmoid'))


model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
modelann = model.fit(X_train, y_train, batch_size=32, epochs=40, validation_split=0.2, verbose=1)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import numpy as np

y_prob = model.predict(X_test)
y_pred = (y_prob > 0.5).astype(int)

cmatrex  = confusion_matrix(y_test, y_pred)
accu = accuracy_score(y_test, y_pred)

print("Confusion Matrix:\n", cmatrex)

print("Accuracy :          ", accu)
```
